```python
#Q2
import pandas as pd
def getDepth(direction, pair):
    r = requests.get("https://data.binance.com/api/v3/depth",params=dict(symbol=pair))
    results = r.json()
    frames = {side: pd.DataFrame(data=results[side], columns=["price", "quantity"],dtype=float) for side in [direction]}
    return frames
getDepth("asks", "ETHBUSD")
```

```python
#Q3
def getOrderBook(pair):
    r = requests.get("https://data.binance.com/api/v3/depth", params=dict(symbol=pair))
    results = r.json()
    frames = {side: pd.DataFrame(data=results[side], columns=["price", "quantity"],dtype=float) for side in ["bids", "asks"]}
    return frames
getOrderBook("ETHBUSD")
```

```python
#Q4
from datetime import datetime
def refreshDataCandle(pair, duration):
    d = requests.get('https://data.binance.com/api/v3/klines?symbol='+pair+'&interval='+duration).json()
    L2=[]
    L3=[]
    for i in range (len(d)):
        L2.append(i)
        L2.append(datetime.fromtimestamp(d[i][0]/1000).strftime("%A, %B %d, %Y %I:%M:%S %p"))
        for j in range (1,6) :
            L2.append(d[i][j])
        L3.append(L2)
        L2=[]
    return L3


print(["id","date","Open price""High price","Low price","Close price","Volume"])

Candl = refreshDataCandle('BTCUSDT', '1m')#Only intervals available on binance (1h, 15m, 5m, 1s...)
for i in Candl:
    print(i)
```
